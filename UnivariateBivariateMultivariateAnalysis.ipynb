# Univariate Analysis

```python
df['species'].unique()
```

```python
df_setosa=df.loc[df['species']=='setosa']
df_versicolor=df.loc[df['species']=='versicolor']
df_virginica=df.loc[df['species']=='virginica']
```

```python
df_setosa.head()
```

```python
plt.plot(df_setosa['petal length (cm)'],np.zeros(len(df_setosa['petal length (cm)'])))
plt.plot(df_versicolor['petal length (cm)'],np.zeros(len(df_versicolor['petal length (cm)'])))
plt.plot(df_virginica['petal length (cm)'],np.zeros(len(df_virginica['petal length (cm)'])))
plt.xlabel('petal length (cm)')
plt.show()
```

```python
plt.plot(df_setosa['petal length (cm)'],np.zeros(len(df_setosa['petal length (cm)'])),'o')
plt.plot(df_versicolor['petal length (cm)'],np.zeros(len(df_versicolor['petal length (cm)'])),'o')
plt.plot(df_virginica['petal length (cm)'],np.zeros(len(df_virginica['petal length (cm)'])),'o')
plt.xlabel('petal length (cm)')
plt.show()
```

# Bivariate  Analysis

```python
sns.FacetGrid(df,hue='species',height=5) \
    .map(plt.scatter,'petal length (cm)','petal width (cm)') \
    .add_legend();
plt.show()
```
